# Decoder-Encoder Experiments


I did experiments on a communication channel with one real-valued number and 16 neurons to understand the relative merits of different decoder (and encoder) initialization strategies.

The basic experiment was to feed a linear ramp into the communication channel, and tested the ability of the channel to produce a straight-line output after a burn-in period of 0.4 seconds. In this simple setting, there were significant differences in accuracy between the following three methods:

1. The current decoder-learning procedure (based on `np.linalg.lstsq`).
2. A drop-in decoder-learning procedure based on ridge regression via lbfgs (probably should be a simpler cg method).
3. A more sophisticated encoder-learning algorithm based on backpropagation through the LIF rate approximation.

In terms of RMSE, these methods of increasing cost and sophistication drop the RMSE of a 16-neuron communication channel from almost 0.035 to just under 0.02.

# The RCOND parameter doesn't do much

The motivation for the use of the rcond parameter to the least squares function was to apply some regularization to the least squares fit, but that doesn't seem to be what it does.

```python
import nengo
from functools import partial
reload(nengo)
model = nengo.Model('RCondTest', seed=123)
model.make_node('Input', lambda t: t - 1)
n_neurons = 16
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.least_squares,
                                            rcond=0.0))
p_ls_10 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.least_squares,
                                            rcond=10000.0))

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0
model.probed[2] = p_ls_10

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='ls')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='ls0')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_10)].probe], label='ls10')
legend(loc='lower right')
```

As you can see in the plot above, the `least_squares` decoder solver with various very different values of rcond seem to produce exactly the same response curves.  Large amounts of regularization should have made the curve be straight and flat, which doesn't happen at all in the plot above.

# Good-old fashioned ridge regression

I thought that the rcond and lstsq method were just doing ridge regression, but evidently not exactly. So I implemented a new ridge-regression solver based on scipy's bfgs wrapper, called `nengo.decoders.ridge_regression`.  The following scripts show that it behaves as expected and provides a better regularization strategy for decoders.

```python
model = nengo.Model('RCond-vs-RidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.ridge_regression,
                                            l2_penalty=0.0))
p_ls_10 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.ridge_regression,
                                            l2_penalty=1000.0))

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0
model.probed[2] = p_ls_10

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='rcond solver')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='RR w no regularization')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_10)].probe], label='RR w huge regularization')
legend(loc='lower right')
```

There we go: the underfitting and overfitting behaviour that we expect.  With very strong regularization, the response is flat near 0, and with no regularization, we have a very eratic line.

## Auto-regularization

Both `least_squares` and `ridge_regression` have a regularization parameter which is not obvious to set a-priori.  For now I wrote a simple loop called `auto_ridge_regression` that provides relatively quick-and-not-too-dirty algorithm for choosing the regularization strength by cross-validation. It takes about 10x as long to run as the `ridge_regression` function, but it provides, I think, really good decoders. (A cleverer implementation could maybe speed this up so it's only around 5x faster than `ridge_regression` by doing a bracket search.) Anyway, I'm going to use this `auto_ridge_regression` from now on in this notebook as the "ridge regression" strategy.


```python
model = nengo.Model('AutoRidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=nengo.decoders.auto_ridge_regression)

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='rcond solver')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='RR w no regularization')
legend(loc='lower right')
```

```python

coefs, residuals_lstsq, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.probe_outputs[sim.model.memo[id(p_ls)].probe]).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using original decoders: coefs=%s resid=%s" % (coefs, residuals_lstsq)

coefs, residuals_arr, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe]).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using ridge-regression decoders: coefs=%s resid=%s" % (coefs, residuals_arr)
```

Obviously both decoder calculations do a similar job, but the auto-ridge regression residuals total 1.34 instead of 1.7. That means that the decoders from `auto_ridge_regression` are better able to decode a straight line input to a straight line output (the slight temporal shift to the right is a normal part of the filtering process).

# Backprop for tuning encoders?

A more computationally costly but sophisticated approach is to backpropagate the decoder error to the encoder values through a rate approximation to the LIF curve. This is done using hacky logic in the `encoders_by_backprop` function. Currently it is configured to do 200 gradient descent steps with fixed decoders and then 200 more steps optimizing both encoders and decoders, but I haven't tried to quantify the shape of the learning curve. Maybe fewer steps would do almost as well, and maybe more steps would do just as much good. Open question.

```python
from nengo.tests.test_backprop import encoders_by_backprop
model = nengo.Model('AutoRidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_arr = model.probe('A', filter=0.03,
                     decoder_solver=nengo.decoders.auto_ridge_regression)
for obj in model.objs.values():
    if isinstance(obj, nengo.objects.Ensemble) and obj.encoders is None:
        encoders_by_backprop(obj, dt=0.001, l2_penalty=1.0, maxfun_nodec=200, maxfun_all=200)
sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.data('A'), label='ARR w backprop')
legend(loc='lower right')

coefs, residuals_bp, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.data('A')).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using ridge-regression decoders and backprop: coefs=%s resid=%s" % (coefs, residuals_bp)
```

# Conclusion

The current way of calculating decoders can be improved by using a simple ridge regression method, and improved furthermore by tuning the encoders as well. The following bar plot shows the RMSE of the communication channel by these methods.

# Follow up: How many neurons do we save by using backprop?

Backprop for choosing encoders can be seen as a build-time investment as opposed to a runtime cost of simply running more neurons to get a certain level of accuracy.  How many neurons does it save?

```python
import nengo
from functools import partial
model = nengo.Model('How many neurons do we save?', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(32), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.auto_ridge_regression))
#                                            max_l2_penalty=10000,
#                                            maxfun=500))

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='rcond solver')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='RR w no regularization')
legend(loc='lower right')
coefs, residuals_lstsq, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.probe_outputs[sim.model.memo[id(p_ls)].probe]).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using original decoders: coefs=%s resid=%s" % (coefs, residuals_lstsq)

coefs, residuals_arr, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe]).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using ridge-regression decoders: coefs=%s resid=%s" % (coefs, residuals_arr)
```

This first stab-in-the-dark follow-up suggests that with about twice as many neurons, the current lstsq method can do about as well as the backprop encoders, and with twice as many [random] encoders, the ridge-regression decoders can do better. It's not clear how to extrapolate this relationship to larger networks and other decoded functions.

# Follow-up:  multiplication (A slightly harder function)

Encoders are more important when you're trying to represent more complicated functions... one function that comes up a lot in NEF models which isn't particularly well handled by random encoders is scalar multiplication.  The following two code snippets do scalar multiplication with both decoder-learning algorithms both with and without backprop.


```python
import nengo
from nengo.tests.test_backprop import encoders_by_backprop

def inputs(t):
    return np.asarray([(t + 5) / 15.0 * np.sin(t), (t + 5) / 15. * np.cos(t)]).T

def find_min_mse(a, b):
    assert len(a) == len(b)
    rmses = [np.sqrt(np.sum((a[:-offset] - b[offset:]) ** 2) / (len(a) - offset))
            for offset in range(1, 100)]
    return min(rmses)

model = nengo.Model('Mult', seed=123)
model.make_node('Input', inputs)
model.make_ensemble('A', nengo.LIF(24), dimensions=2, seed=123)
model.make_ensemble('B', nengo.LIF(50), dimensions=1, seed=123)
model.make_ensemble('C', nengo.LIF(50), dimensions=1, seed=123)
model.connect('Input', 'A')
model.connect('A', 'B', function=lambda x: x[0] * x[1])
model.connect('A', 'C', function=lambda x: x[0] * x[1],
              decoder_solver=nengo.decoders.auto_ridge_regression)
model.probe('B', filter=0.03)
model.probe('C', filter=0.03)
        
sim = model.simulator(dt=0.001)
sim.run(10)
t = sim.data(model.t).ravel()
inputs_t = inputs(t)
plot(t, inputs_t[:,0] * inputs_t[:, 1])
plot(t, sim.data('B'))
plot(t, sim.data('C'))

mult_lstsq_rmse = find_min_mse(inputs_t[:,0] * inputs_t[:, 1], sim.data('B').ravel())
mult_arr_rmse = find_min_mse(inputs_t[:,0] * inputs_t[:, 1], sim.data('C').ravel())
print 'lstsq  RMSE', mult_lstsq_rmse
print 'autoRR RMSE', mult_arr_rmse
```

```python
#for obj in model.objs.values():
#    if isinstance(obj, nengo.objects.Ensemble) and obj.encoders is None:
encoders_by_backprop(model.get('A'),
                     dt=0.001,
                     l2_penalty=1.0,
                     maxfun_nodec=500,
                     maxfun_all=500,
                     ) # -- fun to play with !?
        
sim = model.simulator(dt=0.001)
sim.run(10)
t = sim.data(model.t).ravel()
inputs_t = inputs(t)
plot(t, inputs_t[:,0] * inputs_t[:, 1])
plot(t, sim.data('B'))
plot(t, sim.data('C'))

mult_lstsq_rmse_bp = find_min_mse(inputs_t[:,0] * inputs_t[:, 1], sim.data('B').ravel())
mult_arr_rmse_bp = find_min_mse(inputs_t[:,0] * inputs_t[:, 1], sim.data('C').ravel())
print 'lstsq  RMSE', mult_lstsq_rmse_bp
print 'autoRR RMSE', mult_arr_rmse_bp
```

```python
bar([0, 1, 2, 3], [mult_lstsq_rmse, mult_arr_rmse, mult_lstsq_rmse_bp, mult_arr_rmse_bp])
xticks([.4, 1.4, 2.4, 3.4], ['lstsq', 'auto-r-r', 'lstsq-backprop', 'arr-backprop'])
ylabel('RMSE of mult and channel')
print 'Comparison of encoder/decoder accuracies for different algorithms'
```

It appears that for this multiplication task, backprop doesn't work any miracles, but it helps some.  Using `auto_ridge_regression` does a lot of good by comparison, relative to the lstsq call.  All in all, this accuracy is pretty good for a 20-neuron multiplier.
