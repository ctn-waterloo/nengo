# The RCOND parameter doesn't do much

The motivation for the use of the rcond parameter to the least squares function was to apply some regularization to the least squares fit, but that doesn't seem to be what it does.

```python
import nengo
from functools import partial
reload(nengo)
model = nengo.Model('RCondTest', seed=123)
model.make_node('Input', lambda t: t - 1)
n_neurons = 16
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.least_squares,
                                            rcond=0.0))
p_ls_10 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.least_squares,
                                            rcond=10000.0))

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0
model.probed[2] = p_ls_10

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='ls')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='ls0')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_10)].probe], label='ls10')
legend(loc='lower right')
```

As you can see in the plot above, the `least_squares` decoder solver with various very different values of rcond seem to produce exactly the same response curves.  Large amounts of regularization should have made the curve be straight and flat, which doesn't happen at all.

# Good-old fashioned ridge regression



```python
model = nengo.Model('RCond-vs-RidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.ridge_regression,
                                            l2_penalty=0.0))
p_ls_10 = model.probe('A', filter=0.03,
                     decoder_solver=partial(nengo.decoders.ridge_regression,
                                            l2_penalty=1000.0))

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0
model.probed[2] = p_ls_10

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='rcond solver')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='RR w no regularization')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_10)].probe], label='RR w huge regularization')
legend(loc='lower right')
```

There we go: the underfitting and overfitting behaviour that we expect.  With very strong regularization, the response is flat near 0, and with no regularization, we have a very eratic line.

To get a nice smooth curve, you want to regularize... just right, which you can do by cross-validation.

```python
model = nengo.Model('AutoRidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_ls = model.probe('A', filter=0.03,
                   decoder_solver=nengo.decoders.least_squares)
p_ls_0 = model.probe('A', filter=0.03,
                     decoder_solver=nengo.decoders.auto_ridge_regression)

# hack to get multiple probes of the same thing to build
model.probed[0] = p_ls
model.probed[1] = p_ls_0

sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls)].probe], label='rcond solver')
plot(t, sim.probe_outputs[sim.model.memo[id(p_ls_0)].probe], label='RR w no regularization')
legend(loc='lower right')
```

# Backprop for tuning encoders?

Let's have a look...

```python
from nengo.tests.test_backprop import encoders_by_backprop
model = nengo.Model('AutoRidgeRegression', seed=123)
model.make_node('Input', lambda t: t - 1)
model.make_ensemble('A', nengo.LIF(n_neurons), dimensions=1, seed=123)
model.connect('Input', 'A')
p_arr = model.probe('A', filter=0.03,
                     decoder_solver=nengo.decoders.auto_ridge_regression)
for obj in model.objs.values():
    if isinstance(obj, nengo.objects.Ensemble) and obj.encoders is None:
        encoders_by_backprop(obj, dt=0.001, l2_penalty=1.0)
sim = model.simulator(dt=0.001)
sim.run(2)
t = sim.data(model.t)
plot(t, t - 1)
plot(t, sim.data('A'), label='ARR w backprop')
legend(loc='lower right')

coefs, residuals_bp, rank, sv, rcond = polyfit(
        np.asarray(t).ravel()[400:],
        np.asarray(sim.data('A')).ravel()[400:],
        deg=1,
        full=True)
print "straightness of line using ridge-regression decoders and backprop: coefs=%s resid=%s" % (coefs, residuals_bp)
```

The residual error using encoders specialized for the job is only 0.69. If accuracy is important, finetuning the encoders can improve signal fidelity even in the basic communication channel.  Summary of residual errors is as follows:
